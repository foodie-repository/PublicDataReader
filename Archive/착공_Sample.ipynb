### 단독/다세대/연립

```python
import plotly.express as px

# '계'와 '아파트'를 제외한 나머지 선택
exclude_types = ['계(다가구동수기준)', '아파트']
df_filtered = df_sudo[~df_sudo['분류값명2'].isin(exclude_types)]

# 주택유형별로 그룹화하여 합계 계산 (지역 통합)
df_grouped = df_filtered.groupby(['수록시점', '분류값명2'], as_index=False)['수치값'].sum()

fig = px.line(df_grouped,
            x="수록시점",
            y="수치값",
            color='분류값명2')

fig.update_layout(
    title=dict(
        text='<b>수도권 단독/다세대/연립 착공실적</b>',
        x=0.28,
        y=0.87,
        font=dict(family="Arial", size=25, color="black")
    ),
    hoverlabel=dict(bgcolor='white', font_size=15),
    hovermode="x unified",
    template='plotly_white',
    xaxis_tickangle=90,
    yaxis_tickformat=',',
    legend=dict(orientation='h', xanchor="center", x=0.85, y=1.1),
    legend_title_text='주택유형',
    xaxis=dict(title_text='기준년월', title_font=dict(color='gray')),
    yaxis=dict(title_text='착공실적', title_font=dict(color='gray'))
)

fig.show()
```
